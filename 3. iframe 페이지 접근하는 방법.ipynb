```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs
header_option = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'
}
```

```python
finance_url='https://finance.naver.com/marketindex/'
res=req.get(finance_url,headers=header_option)
html=bs(res.text,'lxml')
```

```python
#1.iframe 요소 접근하기
currency_url='https://finance.naver.com'+html.select_one('iframe')['src']
print(currency_url)
```

```python
#환율데이터를 가진 웹 페이지 요청
c_res=req.get(currency_url,headers=header_option)
html=bs(c_res.text,'lxml')
```

```python
#환율명 가져오기
currencys = html.select('td.tit>a')
```

```python
#매매기준율 가져오기
sales = html.select('td.sale')
display(sales)
```

```python
#현찰 구매가,현찰판매가,송금,수취 가져오기
cash_buy = html.select('td.sale+td')
cash_sell = html.select('td.sale+td+td')
sending = html.select('td.sale+td+td+td')
receiving = html.select('td.sale+td+td+td+td')
```

```python
cb_list = []
for i in cash_buy :
    cb_list.append(i.text.replace('\t','').replace('\n',''))
display(cb_list)
```

```python
cs_list = []
s_list = []
r_list = []
for i in cash_sell :
    cs_list.append(i.text.replace('\t','').replace('\n',''))
for i in sending :
    s_list.append(i.text.replace('\t','').replace('\n',''))
for i in receiving :
    r_list.append(i.text.replace('\t','').replace('\n',''))
```

```python
currency_list = []
sales_list = []
for i in currencys :
    currency_list.append(i.text.replace('\t','').replace('\n',''))
for i in sales :
    sales_list.append(i.text)
```

```python
#2 통화명,매매기준율,현찰구매가,현찰판매가,송금,수취 데이터프레임으로
currency_table = {
    '통화명' : currency_list,
    '매매기준율' : sales_list,
    '현찰구매가' : cb_list,
    '현찰판매가' : cs_list,
    '송금' : s_list,
    '수취' : r_list
}
ct_df = pd.DataFrame(currency_table)
```
