# requests 라이브러리
- 접근할 때 웹 페이지의 데이터를 요청/응답 받기 위한 라이브러리

```python
import requests as req
```

```python
#1.수집할 웹 페이지의 주소 정의
url = 'http://www.naver.com'
```

```python
#2.라이브러리를 이용해서 웹 페이지 요청
#http 상태코드
# - 200: 성공
# - 400: 클라이언트 오류
# - 500: 서버 오류
res=req.get(url)
res
```

```python
#3. 웹 페이지 확인하기
res.text
```

# 네이버 뉴스 헤드라인 URL 가져오기 실습

```python
nh_url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105'
```

```python
#RemoteDisconnected 오류 해결방법
# - 브라우저로 요청을 보냈다라는 것을 서버에서 인식할 수 있도록
# user-agent 값을 구성하고 보내줘야 한다!
header_option = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'
}
nh_res=req.get(nh_url,headers=header_option)
nh_res
```

```python
nh_res.text
```

```python
nh_html=bs(nh_res.text,'lxml')
nh_html
```

```python
nh_rinkEle=nh_html.select('a.sh_text_headline')
```

```python
nh_rinkEle
```

```python
#요소의 속성값 가져오기
#- 요소객체[속성명]
nh_rinkEle[0]['href']
```

```python
#헤드라인 기사의 url를 가져온 후 h_url_list에 추가하기
h_url_list = []
for i in nh_rinkEle :
    h_url_list.append(i['href'])
    
display(h_url_list)
```

```python
for url in h_url_list : 
    res = req.get(url, headers=header_option)
    html=bs(res.text,'lxml')

    title=html.select_one('#title_area>span')
    content=html.select_one('#dic_area')

    print('제목:',title.text)
    print('내용:',content.text.strip().replace('\n',''))
    print()
```
