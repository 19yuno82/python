```python
from selenium import webdriver as wb
#html 요소에 접근하기 위한 방법을 제공하는 라이브러리
from selenium.webdriver.common.by import By
#키보드 값을 제공하는 라이브러리
from selenium.webdriver.common.keys import Keys
```

```python
#chromedriver.exe 파일 실행하여 테스트 브라우저 띄우기
driver = wb.Chrome()
```

```python
#테스트 브라우저에 접근할 URL주소 전달
#=> 접근한 페이지의 내용을 응답받아서 브라우저에 출력
driver.get('https://www.naver.com')
```

```python
#웹 페이지 내용 문자열 출력
driver.page_source
```

```python
#html 요소 접근하기
searchInput = driver.find_element(by='id',value='query')
```

```python
#검색창에 키워드 전달하기
searchInput.send_keys('장난감\n')
```

```python
#검색창에 키워드 전달 후 검색버튼 클릭하기
searchBtn=driver.find_element(By.CLASS_NAME, value='btn_search')

searchInput.send_keys('피규어')
searchBtn.click()
```

```python
#브라우저 뒤로 가기
driver.back()
```

```python
#스크롤 내리기
body = driver.find_element(By.TAG_NAME, value='body')
body.send_keys(Keys.PAGE_DOWN)
```
